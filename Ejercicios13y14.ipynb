```python
#Ejercicio 13

def leer_entero(mensaje):
    while True:
        try:
            valor = int(input(mensaje))
            return valor
        except ValueError:
            print("Por favor, ingresa un número entero válido.")

def generar_matriz(filas, columnas):
    matriz = []
    print("Ingrese los elementos de la matriz:")
    for i in range(filas):
        fila = []
        for j in range(columnas):
            elemento = leer_entero(f"Ingrese el elemento en la fila {i+1}, columna {j+1}: ")
            fila.append(elemento)
        matriz.append(fila)
    return matriz

def imprimir_matriz(matriz):
    for fila in matriz:
        print(fila)

def main():
    filas = leer_entero("Ingrese el número de filas de la matriz: ")
    columnas = leer_entero("Ingrese el número de columnas de la matriz: ")

    matriz = generar_matriz(filas, columnas)

    print("Matriz resultante:")
    imprimir_matriz(matriz)

if __name__ == "__main__":
    main()
```

```python
#Ejercicio 14
def leer_entero(mensaje):
    while True:
        try:
            valor = int(input(mensaje))
            return valor
        except ValueError:
            print("Por favor, ingresa un número entero válido.")

def leer_matriz(filas, columnas, nombre):
    matriz = []
    print(f"Ingrese los elementos de la matriz {nombre}:")
    for i in range(filas):
        fila = []
        for j in range(columnas):
            elemento = leer_entero(f"Ingresa el elemento en la fila {i+1}, columna {j+1}: ")
            fila.append(elemento)
        matriz.append(fila)
    return matriz

def imprimir_matriz(matriz, nombre):
    print(f"Matriz {nombre}:")
    for fila in matriz:
        print(fila)

def suma_matrices(matrices):
    resultado = matrices[0]
    for i in range(1, len(matrices)):
        for j in range(len(resultado)):
            for k in range(len(resultado[0])):
                resultado[j][k] += matrices[i][j][k]
    return resultado

def main():
    cantidad = leer_entero("Ingrese la cantidad de matrices que desea sumar: ")

    if cantidad < 2:
        print("Debe ingresar al menos dos matrices para sumar.")
        return

    filas = leer_entero("Ingrese el número de filas de las matrices: ")
    columnas = leer_entero("Ingrese el número de columnas de las matrices: ")

    matrices = []
    for i in range(cantidad):
        nombre = chr(ord('A') + i)
        matriz = leer_matriz(filas, columnas, nombre)
        matrices.append(matriz)
        imprimir_matriz(matriz, nombre)

    resultado = suma_matrices(matrices)

    print("Resultado de la suma de las matrices:")
    for i in range(cantidad):
        nombre = chr(ord('A') + i)
        if i > 0:
            print("+")
        imprimir_matriz(matrices[i], nombre)
    print("=")
    imprimir_matriz(resultado, "Resultado")

if __name__ == "__main__":
    main()
```
